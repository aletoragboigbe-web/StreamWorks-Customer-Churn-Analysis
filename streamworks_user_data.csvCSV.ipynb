```python
import pandas as pd

DATA_PATH = r"C:\Users\Aletor\Desktop\streamworks_user_data.csv"
df = pd.read_csv("streamworks_user_data.csv")
df.head()
```

```python
# Cell 3 - Quick structure & missing values
print("=== dtypes ===")
print(df.dtypes)
print("\n=== info() ===")
df.info()
print("\n=== numeric describe() ===")
display(df.describe().T)
print("\n=== Missing values per column ===")
print(df.isnull().sum().sort_values(ascending=False))
```

```python
# Cell 4 - Convert date columns and create tenure_days + is_loyal
date_cols = [c for c in df.columns if 'date' in c.lower() or 'signup' in c.lower() or 'last_active' in c.lower()]
print("Found potential date columns:", date_cols)
```

```python
# Convert to datetime safely
for c in date_cols:
    df[c] = pd.to_datetime(df[c], errors='coerce')
```

```python
# Create tenure_days if both signup and last active exist
if 'signup_date' in df.columns and 'last_active_date' in df.columns:
    df['tenure_days'] = (df['last_active_date'] - df['signup_date']).dt.days
    print("Created tenure_days.")
else:
    print("signup_date or last_active_date not both present; check your column names.")
```

```python
# Create is_loyal flag (>180 days)
if 'tenure_days' in df.columns:
    df['is_loyal'] = df['tenure_days'] > 180
    df['is_loyal'] = df['is_loyal'].astype(int)  # 1 or 0
```

```python
# Simple missing value strategy
# Numeric -> fill median; Categorical -> fill 'Unknown'
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(include=['object','category']).columns.tolist()

for c in num_cols:
    if df[c].isnull().any():
        df[c] = df[c].fillna(df[c].median())

for c in cat_cols:
    if df[c].isnull().any():
        df[c] = df[c].fillna('Unknown')

print("Filled missing values for numeric (median) and categorical (Unknown).")
```

```python
# Feature engineering: watch_per_fee_ratio, heavy_mobile_user, age bins
# Only create if columns exist; adapt names to your file
if 'average_watch_hours' in df.columns and 'monthly_fee' in df.columns:
    df['watch_per_fee_ratio'] = df['average_watch_hours'] / df['monthly_fee'].replace(0, np.nan)
    df['watch_per_fee_ratio'] = df['watch_per_fee_ratio'].fillna(0)
```

```python
# heavy_mobile_user example: if you have a device column
device_col = None
for c in df.columns:
    if 'device' in c.lower():
        device_col = c
        break

if device_col and 'average_watch_hours' in df.columns:
    med = df['average_watch_hours'].median()
    df['heavy_mobile_user'] = ((df[device_col].astype(str).str.lower() == 'mobile') &
                               (df['average_watch_hours'] > med)).astype(int)
```

```python
# Age bins example (if age exists)
if 'age' in df.columns:
    bins = [0, 17, 24, 34, 44, 54, 64, 120]
    labels = ['<18','18-24','25-34','35-44','45-54','55-64','65+']
    df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
# Encode categorical columns for modeling (LabelEncoder for simplicity)
enc = LabelEncoder()
encoded_cols = []

for c in ['gender','subscription_type','country','received_promotions','referred_by_friend']:
    if c in df.columns:
        df[c + '_enc'] = enc.fit_transform(df[c].astype(str))
        encoded_cols.append(c + '_enc')

print("Encoded columns (for models):", encoded_cols)
```

```python
# Correlation matrix & heatmap (numeric)
num_df = df.select_dtypes(include=[np.number])
corr = num_df.corr()
display(corr)

# Simple heatmap using matplotlib (one-plot rule)
plt.figure(figsize=(10,8))
plt.imshow(corr, interpolation='nearest', aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.index)), corr.index)
plt.title("Correlation matrix (numeric)")
plt.tight_layout()
plt.show()
```

```python
# Statistical tests: Chi-square and t-test
# Ensure we have a churn column; adapt name if needed
churn_col = None
for c in df.columns:
    if 'churn' in c.lower() or 'cancel' in c.lower():
        churn_col = c
        break

if churn_col is None:
    raise ValueError("No churn column found. Rename your churn column (e.g., is_churned).")
```

```python
# Create binary churn_flag (0/1)
df['churn_flag'] = df[churn_col].apply(lambda x: 1 if str(x).strip().lower() in ['1','true','yes','y','churned','cancelled','cancel'] or x==1 else 0)

print("Churn distribution:")
print(df['churn_flag'].value_counts())
```

```python
# Chi-square for gender, received_promotions, referred_by_friend (if present)
for cat in ['gender','received_promotions','referred_by_friend']:
    if cat in df.columns:
        ct = pd.crosstab(df[cat].astype(str), df['churn_flag'])
        if ct.shape[0] > 1:
            chi2, p, dof, ex = chi2_contingency(ct)
            print(f"Chi-square for {cat}: chi2={chi2:.3f}, p={p:.3f}")
        else:
            print(f"{cat}: not enough categories for chi-square")
```

```python
# t-test for average_watch_hours between churned and retained
if 'average_watch_hours' in df.columns:
    churn_watch = df[df['churn_flag']==1]['average_watch_hours']
    ret_watch = df[df['churn_flag']==0]['average_watch_hours']
    tstat, pval = ttest_ind(churn_watch, ret_watch, equal_var=False, nan_policy='omit')
    print(f"T-test average_watch_hours: t={tstat:.3f}, p={pval:.3f}")
```

```python
# Visualizations: boxplot and churn rate by subscription type
# Boxplot of watch hours by churn
if 'average_watch_hours' in df.columns:
    df.boxplot(column='average_watch_hours', by='churn_flag')
    plt.suptitle('')
    plt.title('Average watch hours by churn_flag')
    plt.xlabel('churn_flag (0=retained, 1=churn)')
    plt.ylabel('average_watch_hours')
    plt.show()
```

```python
# Churn rate by subscription_type
if 'subscription_type' in df.columns:
    churn_by_sub = df.groupby('subscription_type')['churn_flag'].mean().sort_values(ascending=False)
    churn_by_sub.plot(kind='bar')
    plt.ylabel('Churn rate')
    plt.title('Churn rate by subscription_type')
    plt.show()
```

```python
# Logistic Regression to predict churn
# Select features (pick numeric + encoded categorical)
features = []
for f in ['average_watch_hours','tenure_days','monthly_fee','is_loyal','watch_per_fee_ratio','heavy_mobile_user']:
    if f in df.columns:
        features.append(f)
# add up to 4 encoded categorical columns
features += [c for c in encoded_cols][:4]

print("Features used:", features)

X = df[features].fillna(0)
y = df['churn_flag']
```

```python
# Train/Test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
# Scale
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
# Fit model
log = LogisticRegression(max_iter=1000)
log.fit(X_train_s, y_train)
```

```python
# Predict & evaluate
y_pred = log.predict(X_test_s)
y_proba = log.predict_proba(X_test_s)[:,1]
```

```python
print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification report:")
print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))
```

```python
# Coefficients (sorted by absolute importance)
coef_df = pd.DataFrame({
    'feature': features,
    'coef': log.coef_[0]
}).assign(abs_coef=lambda d: d['coef'].abs()).sort_values('abs_coef', ascending=False)
display(coef_df)
```

```python
# Cell 12 - ROC curve plot
fpr, tpr, _ = roc_curve(y_test, y_proba)
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, y_proba):.3f}")
plt.plot([0,1],[0,1], '--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve - Logistic Regression')
plt.legend()
plt.show()
```
